```python
opc=5
def X ():
 plt.plot(data_ventas["Xignux"], label="Xignux") #.plot(kind="line", label="Xignux")
 #fig=ax.get_figure()
 #fig.savefig("grafica_análisis_ventas")
 return #fig
def V ():
 ax=data_ventas["Viakable"].plot(kind="line" ,label="Viakable")
 fig=ax.get_figure()
 #fig.savefig("grafica_análisis_ventas")
 return fig
def P ():
 ax=data_ventas["Prolec"].plot(kind="line",label="Prolec")
 fig=ax.get_figure()
 #fig.savefig("grafica_análisis_ventas")
 return fig
def Q ():
 ax=data_ventas["Qualtia"].plot(kind="line",label="Qualtia")
 fig=ax.get_figure()
 #fig.savefig("grafica_análisis_ventas")
 return fig
def Vy ():
 ax=data_ventas["Vydsa"].plot(kind="line",label="Vydsa")
 fig=ax.get_figure()
 #fig.savefig("grafica_análisis_ventas")
 return fig

while opc < 6:  
    print ("*** Seleccione la gráfica a incluir en el análisis**")
    print (" 1.Xignux\n 2. Viakable\n 3. Prolec \n 4. Qualtia \n 5. Vydsa")
    opc = int(input("Gráfica a incluir(6 para salir)?")) 
    if opc == 1:
        print("Se agregó:---Xignux---",opc)
        X()
    elif opc == 2:
        print("Se agregó:--- Viakable---",opc)
        V()
    elif opc == 3:
        print("Se agregó:---Prolec---",opc)
        P()
    elif opc == 4:
        print("Se agregó---Qualtia",opc)
        Q()
    elif opc == 5:
        print("Se agrego---Vydsa---",opc)
        Vy()
    else:
        plt.ylabel("Ventas")
        plt.xlabel("Año/Mes")
        plt.xticks(rotation=65)
        plt.title("Pronóstico de Ventas")
        plt.legend(loc="upper left")
        plt.savefig("grafica_análisis_ventas")
        print ("La gráfica es:")
```
